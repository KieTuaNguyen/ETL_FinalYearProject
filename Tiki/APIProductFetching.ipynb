```python
import requests
import json
import pandas as pd
```

```python
URL = "https://tiki.vn/api/v2/products/270736028"

HEADERS = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36 Edg/121.0.0.0",
    "From": "",
    "af-ac-enc-dat": "",
    "x-api-source": "pc",
}
```

```python
response = requests.get(URL, headers=HEADERS)

if response.status_code == 200:
    data = response.json()
    print("API Response:", json.dumps(data, indent=4))
else:
    print("Failed to fetch data:", response.status_code)
```

```python
# Count the number of datasets in "data" list
num_attr = len(data)
print("Number of attributes in the dataset:", num_attr)
```

```python
all_values = [key for key, value in data.items()]
print("All key from the JSON response:")
print(json.dumps(all_values, indent=4))
```

```python
# Create a DataFrame from the JSON data
df = pd.json_normalize(data)

# Flatten all columns containing nested JSON structures
for column in df.columns:
    if isinstance(df[column][0], list) or isinstance(df[column][0], dict):
        df = pd.concat([df.drop([column], axis=1), 
                        pd.json_normalize(df[column].explode())
                        .groupby(level=0).agg(list)], axis=1)
```

```python
# Save the DataFrame to a CSV file with UTF-8 encoding
df.to_csv('Test.csv', index=False, encoding='utf-8-sig')
```
