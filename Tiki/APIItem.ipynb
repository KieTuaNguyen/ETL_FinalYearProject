```python
import pandas as pd
import requests
import math
```

```python
category_id = 625
base_url = "https://tiki.vn/api/personalish/v1/blocks/listings"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36 Edg/121.0.0.0",
    "Accept-Language": 'en-US,en;q=0.9',
    "Accept-Encoding": "gzip, deflate, br, zstd",
    "Referer": "https://tiki.vn/",
    "From": "",
    "af-ac-enc-dat": "",
    "x-api-source": "pc"
}
params = {"category": category_id, "page": 1}
```

```python
# Make initial request to get total count
response = requests.get(base_url, headers=headers, params=params)
data = response.json()
data
```

```python
total = data["paging"]["total"]
per_page = data["paging"]["per_page"]
last_page = data["paging"]["last_page"]

# Calculate number of pages needed
num_pages = math.ceil(total / per_page)
print(f"Total pages fetched: {num_pages}")
```

```python
# Fetch data from each page
all_items = []
for page in range(1, num_pages+1):
    params = {"category": category_id, "page": page} 
    response = requests.get(base_url, headers=headers, params=params)
    data = response.json()
    
    for item in data["data"]:
        id = item["id"]
        name = item["name"]
        
        all_items.append([id, name])

print(f"Total items fetched: {len(all_items)}")
df = pd.DataFrame(all_items, columns=["Item_ID", "Item_Name"])
```
